```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time # スリープを使うために必要
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import random
```

```python
driver.implicitly_wait(5)
url = "https://kuji.rakuten.co.jp/" #楽天くじ公式サイトのurl
driver.get(url)
# 1個目のラッキーくじを選択
path1 = "/html/body/div/main/section[2]/ul/li[1]/a"
btn = driver.find_element(By.XPATH, path1)
btn.click()
time.sleep(5)
try:
  # ログイン情報入力
  textbox = driver.find_element(By.ID,"loginInner_u")
  textbox.send_keys("") #楽天アカウントのIDを入力
  time.sleep(5)
  textbox = driver.find_element(By.ID,"loginInner_p")
  textbox.send_keys("") #楽天アカウントのパスワードを入力
  time.sleep(5)
  # ログインボタンを押す
  btn = driver.find_element(By.XPATH,"/html/body/div[2]/div/div/div[1]/div/form/div/p[1]/input")
  btn.click()
  time.sleep(5)
except NoSuchElementException:
  print("- NoSuchElementException発生 - 次の処理へ")
try:
 # くじ「Start」をクリック 
 btn = driver.find_element(By.XPATH,"//*[@id='entry']")
 btn.click()
 time.sleep(30)
except NoSuchElementException:
 print("- NoSuchElementException発生 - 次の処理へ")
 time.sleep(5)

for num in range(2, 18):
  print(str(num-1)+"回")
  path2 = "/html/body/div/main/section[2]/ul/li[" + str(num) + "]/a"

  # 楽天ラッキーくじ トップページに移動
  driver.get('https://kuji.rakuten.co.jp/')
  time.sleep(15)
  try:
    # ラッキーくじを順番に選択
    btn = driver.find_element(By.XPATH,path2)
    btn.click()
    t = random.randint(10,15)
    time.sleep(t)
    
    try:
    # くじ「Start」をクリック
      btn = driver.find_element(By.XPATH,"//*[@id='entry']")
      btn.click()

      t = random.randint(30,40)
      time.sleep(t)
    except NoSuchElementException:
      print("- NoSuchElementException発生 - 次の処理へ")
      t = random.randint(10,25)
      time.sleep(t)
    except Exception:
      print("-Exception発生 - 次の処理へ")

  except NoSuchElementException:
    print("- NoSuchElementException発生 - 次の処理へ")
    break

#ブラウザを閉じる
driver.close()
```
